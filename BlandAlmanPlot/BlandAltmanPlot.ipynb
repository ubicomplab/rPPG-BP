```python
import pandas as pd
import numpy as np
import BlandAltmanPy
```

## Window size of 10

```python
# Load example data file:
window_size = 10
finger_hr = np.load(f'../results/finger_hr_all_ws_{window_size}.npy')
face_hr = np.load(f'../results/face_hr_all_ws_{window_size}.npy')
print('face hr. shape: ', face_hr.shape)
print('finger hr. shape: ', finger_hr.shape)
compare = BlandAltmanPy.BlandAltman(finger_hr,face_hr,averaged=True)
compare.scatter_plot(x_label='Finger PPG HR [bpm]',y_label='Face PPG HR [bpm]')
compare.difference_plot()
compare.print_stats(round_amount = 3)
```
