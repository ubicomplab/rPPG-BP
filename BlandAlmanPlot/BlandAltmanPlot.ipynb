```python
import pandas as pd
import numpy as np
import BlandAltmanPy
```

## Window size of 10

```python
# Load example data file:
window_size = 10
face_hr = np.load(f'../results/face_hr_all_ws_{window_size}.npy')
ecg_hr = np.load(f'../results/ecg_hr_all_ws_{window_size}.npy')
print('face hr. shape: ', face_hr.shape)
print('ecg hr. shape: ', ecg_hr.shape)
compare = BlandAltmanPy.BlandAltman(ecg_hr,face_hr,averaged=True)
compare.scatter_plot(x_label='ECG HR [bpm]',y_label='rPPG HR [bpm]', figure_size=(5, 5), file_name=f'ScatterPlot{window_size}.pdf')
compare.difference_plot(figure_size=(5, 5), file_name=f'BlandAltman{window_size}.pdf')
compare.print_stats(round_amount = 3)
```
